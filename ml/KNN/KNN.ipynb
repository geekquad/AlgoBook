# KNN Supervised Machine Learning Algorithm 

### Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
```

### Loading Data

```python
data = np.load('./mnist_train_small.npy')
```

```python
X = data[:, 1:]
y = data[:, 0]
```

```python
X.shape
```

```python
y.shape
```

```python
plt.imshow(X[2].reshape(28, 28), cmap='gray')
```

```python
y[2]
```

### Train-Test-Split

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.33, random_state=42)
```

### KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
```

```python
# training the model happens in fit function
# no work during training knn model
model.fit(X_train, y_train)
```

### Prediction using the model

```python
model.predict(X_test[:10])
```

```python
y_test[:10]
```

```python
plt.imshow(X_test[2].reshape(28, 28), cmap='gray')
```

### Accuracy of the model

```python
model.score(X_test[:100], y_test[:100])
```

# Thus accuracy of model is 95%
